```python
import pandas as pd
import numpy as np
```

### Extracting Data


```python
Data=pd.read_csv("be_properties.csv")
Data
```

### Filling Empty Cells

```python
Data = Data.replace('', np.nan).fillna(value=np.nan)
```

```python
Data.info()
```

### Name cleaning


```python
Data['locality'] = Data['locality'].str.replace(r'[^a-zA-Z\s]', '', regex=True).str.title()
```

```python
Data.locality
```

```python
Data.columns
```

### Columns name fixing

```python
cols=['Id', 'Locality', 'Type', 'SubType', 'Price', 'Bedrooms',
       'Living_Area', 'Is_Equiped_Kitchen', 'Is_Furnished', 'Is_Open_Fire', 'Terrace',
       'Terrace_Area', 'Garden', 'Garden_Area', 'Land_Surface',
       'Surface of the plot', 'State', 'Facades', 'Swim_pool']
```

```python
Data.columns=cols
Data
```

```python
Data.Type=Data.Type.str.title()
```

```python
Data
```

```python
Data.SubType.value_counts(ascending=False)
```

### Price type adjustment and fill null


```python
Data.Price=Data.Price.astype(int)
```

```python
Data['Price'] = Data['Price'].fillna(0)  # Replace NaN with 0
Data['Price'].replace([np.inf, -np.inf], 0, inplace=True)  # Replace Inf/-Inf with 0

# Convert to integer
Data['Price'] = Data['Price'].astype(int)
```

```python
Data.Bedrooms.fillna(0).astype(int)
```

### Booleans adjustments


```python
Data[['Is_Equiped_Kitchen', 'Is_Furnished', 'Is_Open_Fire','Terrace']]=Data[['Is_Equiped_Kitchen', 'Is_Furnished', 'Is_Open_Fire','Terrace']].astype(bool)
```

```python
Data.Bedrooms= Data.Bedrooms.fillna(0).astype(int)
```

```python
Data.Garden_Area=Data.Garden_Area.fillna(0)
```
